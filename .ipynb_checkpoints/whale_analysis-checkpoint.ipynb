Evaluate the performance of various algorithmic, hedge, and mutual fund portfolios and compare them against the S&P 500 Index.

```python
# Initial imports
import pandas as pd
import numpy as np
import datetime as dt
from pathlib import Path

%matplotlib inline
```

# Data Cleaning

Read the CSV files into DataFrames and perform necessary data cleaning steps. After cleaning, combine all DataFrames into a single DataFrame.

Files:

* `whale_returns.csv`: Contains returns of some famous "whale" investors' portfolios.

* `algo_returns.csv`: Contains returns from the in-house trading algorithms from Harold's company.

* `sp500_history.csv`: Contains historical closing prices of the S&P 500 Index.

Defining a function to read portfolio csv file

```python
#define a function to read the csv for multiple file
def csv_reader(csv_file_name,index_field):
    #defining the csv path
    csv_path = Path("Resources/"+csv_file_name+".csv")
    #creating dateframe and fixing index
    df_from_csv = pd.read_csv(csv_path,index_col=index_field, infer_datetime_format=True, parse_dates=True)
    return df_from_csv
```

## Whale Returns

Read the Whale Portfolio daily returns and clean the data

```python
# Reading whale returns and making date column as index
whale_returns_df = csv_reader('whale_returns','Date')
# display few records
whale_returns_df.head()
```

```python
# Count nulls
whale_returns_df.isnull().sum()
```

```python
# Drop nulls
whale_returns_df.dropna(inplace=True)
# Count nulls to make sure it got dropped
whale_returns_df.isnull().sum()
```

## Algorithmic Daily Returns

Read the algorithmic daily returns and clean the data

```python
# Reading whale returns and making date column as index
algo_returns_df = csv_reader('algo_returns','Date')
# display few records
algo_returns_df.head()
```

```python
# Count nulls
algo_returns_df.isnull().sum()
```

```python
# Drop nulls
algo_returns_df.dropna(inplace = True)
# Count nulls to make sure it got dropped
algo_returns_df.isnull().sum()
```

## S&P 500 Returns

Read the S&P 500 historic closing prices and create a new daily returns DataFrame from the data. 

```python
# Reading whale returns and making date column as index
sp500_history_df = csv_reader('sp500_history','Date')
# display few records
sp500_history_df.head()
```

```python
# Check Data Types
sp500_history_df.dtypes
```

```python
# Fix Data Types: Removed $ from close and converted it to float from object
sp500_history_df["Close"] = sp500_history_df["Close"].str.replace('$','').astype(float)
# Check Data Types
sp500_history_df.dtypes
```

```python
# Calculate Daily Returns
sp500_return_df = sp500_history_df.pct_change()
# display few records
sp500_return_df.head()
```

```python
# Drop nulls
sp500_return_df.dropna(inplace = True)
# Count nulls to make sure it got dropped
sp500_return_df.isnull().sum()
```

```python
# Rename `Close` Column to be specific to this portfolio.
sp500_return_df = sp500_return_df.rename(columns={"Close" : "sp500_return"})
# display few records
sp500_return_df.head()
```

## Combine Whale, Algorithmic, and S&P 500 Returns

```python
# Join Whale, Algorithmic, and the S&P 500 Returns into a single DataFrame with columns for each portfolio's returns.
combined_return_df = pd.concat([whale_returns_df,algo_returns_df,sp500_return_df],axis="columns",join="inner")
# display few records
combined_return_df.head()
```

```python
# Count nulls
combined_return_df.isnull().sum()
```

```python
# Drop nulls
combined_return_df.dropna(inplace = True)
# Count nulls to make sure it got dropped
combined_return_df.isnull().sum()
```

# Conduct Quantitative Analysis

Calculate and visualize performance and risk metrics for the portfolios.

## Performance Anlysis

#### Calculate and Plot the daily returns.

```python
# Plot daily returns of all portfolios

combined_return_df.plot(xlabel="date",ylabel= "daily return",title="Daily Return Comparison", figsize=(14,7));
```

#### Calculate and Plot cumulative returns.

```python
# Calculate cumulative returns of all portfolios
comb_cum_return_df = (1 + combined_return_df).cumprod()
# display few records
comb_cum_return_df.head()
```

---

```python
# Plot cumulative returns
comb_cum_return_df.plot(xlabel="date",ylabel= "cumulative return",title="Cumulative Return Comparison", figsize=(14,7));
```

## Risk Analysis

Determining the _risk_ of each portfolio:

1. Create a box plot for each portfolio. 
2. Calculate the standard deviation for all portfolios
4. Determine which portfolios are riskier than the S&P 500
5. Calculate the Annualized Standard Deviation

### Creating a box plot for each portfolio


```python
# Box plot to visually show risk
combined_return_df.plot.box(figsize=(14, 7), title='Box Plot of Returns');
```

### Calculating Standard Deviations

```python
# Calculate the daily standard deviations of all portfolios
daily_std_dev = combined_return_df.std()
daily_std_dev
```

### Determine which portfolios are riskier than the S&P 500

```python
# Calculate  the daily standard deviation of S&P 500
sp500_std_dev = combined_return_df["sp500_return"].std()
sp500_std_dev
# Determine which portfolios are riskier than the S&P 500
for column in combined_return_df:
    if sp500_std_dev < combined_return_df[column].std():
        print(f"'{column}' standard deviation is greater than S&P 500, hence it is riskier than the S&P 500")
```

### Calculate the Annualized Standard Deviation

```python
# Calculate the annualized standard deviation (252 trading days)
annaulized_std = combined_return_df.std() * np.sqrt(252)
annaulized_std
```

## Rolling Statistics

Risk changes over time. Analyze the rolling statistics for Risk and Beta. 

1. Calculate and plot the rolling standard deviation for the S&P 500 using a 21-day window
2. Calculate the correlation between each stock to determine which portfolios may mimick the S&P 500
3. Choose one portfolio, then calculate and plot the 60-day rolling beta between it and the S&P 500

### Calculate and plot rolling `std` for all portfolios with 21-day window

```python
# Calculate the rolling standard deviation for all portfolios using a 21-day window
portfolios_rolling_std = combined_return_df.rolling(window=21).std()
# Plot the rolling standard deviation
portfolios_rolling_std.plot(xlabel="date",ylabel= "standard deviation",title="21-day Rolling Std Dev", figsize=(14,7));
```

### Calculate and plot the correlation

```python
# Calculate the correlation
correlation_df = combined_return_df.corr()
# Display de correlation matrix
correlation_df.style.background_gradient()
```

### Calculate and Plot Beta for a chosen portfolio and the S&P 500

```python
# Calculate covariance of a single portfolio
covariance = combined_return_df['SOROS FUND MANAGEMENT LLC'].rolling(window=21).cov(combined_return_df['sp500_return'])
# Calculate variance of S&P 500
variance = combined_return_df['sp500_return'].rolling(window=21).var()
# Computing beta
beta = covariance/variance
beta
# Plot beta trend
beta.plot(figsize=(14,7));
```

## Rolling Statistics Challenge: Exponentially Weighted Average 

An alternative way to calculate a rolling window is to take the exponentially weighted moving average. This is like a moving window average, but it assigns greater importance to more recent observations. Try calculating the [`ewm`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.ewm.html) with a 21-day half-life.

```python
# Use `ewm` to calculate the rolling window
combined_return_df.ewm(halflife=21).std()
```

# Sharpe Ratios
In reality, investment managers and thier institutional investors look at the ratio of return-to-risk, and not just returns alone. After all, if you could invest in one of two portfolios, and each offered the same 10% return, yet one offered lower risk, you'd take that one, right?

### Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot

```python
# Annualized Sharpe Ratios
sharpeRatio = (combined_return_df.mean() * 252) / (combined_return_df.std() * np.sqrt(252))
sharpeRatio
```

```python
# Visualize the sharpe ratios as a bar plot
sharpeRatio.plot(kind="bar",xlabel="portfolio",ylabel= "sharpe ratio",title="Sharpe Ratio Comparison", figsize=(14,7));
```

### Determine whether the algorithmic strategies outperform both the market (S&P 500) and the whales portfolios.

Sharpe Ratio for Algo1 is the highest, hence it is outperforming both the market (S&P 500) and whales portfolios.
Sharpe Ratio for Algo2 is higher compared to market (S&P 500) but lower compared to one of the whale portfolios 'BERKSHIRE HATHAWAY INC', hence it is outperforming the market (S&P 500) but not all the whale portfolios.

## Choose 3-5 custom stocks with at last 1 year's worth of historic prices and create a DataFrame of the closing prices and dates for each stock.

For this demo solution, we fetch data from three companies listes in the S&P 500 index.

* `GOOG` - [Google, LLC](https://en.wikipedia.org/wiki/Google)

* `AAPL` - [Apple Inc.](https://en.wikipedia.org/wiki/Apple_Inc.)

* `COST` - [Costco Wholesale Corporation](https://en.wikipedia.org/wiki/Costco)

```python
# Reading data from goog_historical.csv using function csv_reader defined at the top
goog_historical_df = csv_reader('goog_historical','Trade DATE')
# dropping column symbol and renaming NOCP as portfolio name
goog_historical_df = goog_historical_df.drop(columns = ['Symbol']).rename(columns = {'NOCP': 'goog'})
# Renaming index from Trade DATE to Date to make it easier to merge with portfolio data
goog_historical_df.index.names = ['Date']
# display few records
goog_historical_df.head()
```

```python
# Reading data from aapl_historical.csv using function csv_reader defined at the top
aapl_historical_df = csv_reader('aapl_historical','Trade DATE')
# dropping column symbol and renaming NOCP as portfolio name
aapl_historical_df = aapl_historical_df.drop(columns = ['Symbol']).rename(columns = {'NOCP': 'aapl'})
# Renaming index from Trade DATE to Date to make it easier to merge with portfolio data
aapl_historical_df.index.names = ['Date']
# display few records
aapl_historical_df.head()
```

```python
# Reading data from aapl_historical.csv using function csv_reader defined at the top
cost_historical_df = csv_reader('cost_historical','Trade DATE')
# dropping column symbol and renaming NOCP as portfolio name
cost_historical_df = cost_historical_df.drop(columns = ['Symbol']).rename(columns = {'NOCP': 'cost'})
# Renaming index from Trade DATE to Date to make it easier to merge with portfolio data
cost_historical_df.index.names = ['Date']
# display few records
cost_historical_df.head()
```

```python
# Combine all stocks in a single DataFrame
combined_stock_df = pd.concat([goog_historical_df,aapl_historical_df,cost_historical_df],axis="columns",join="inner")
# display few records
combined_stock_df.head()
```

```python
# Reset Date index
```

```python
# Reorganize portfolio data by having a column per symbol
# display few records
```

```python
# Calculate daily returns
comb_stock_return_df = combined_stock_df.pct_change()
# Drop NAs
comb_stock_return_df.dropna(inplace = True)
# Display sample data
comb_stock_return_df.head()
```

## Calculate the weighted returns for the portfolio assuming an equal number of shares for each stock

```python
# Set weights
weights = [1/3, 1/3, 1/3]
columns = comb_stock_return_df.columns
# Calculate portfolio return
comb_stock_return_df['MY_PORTFOLIO_RETURN'] = comb_stock_return_df[columns].mul(weights).sum(1)
my_portfolio_return_df = comb_stock_return_df.drop(columns = ['goog','aapl','cost'],axis=1)
# Display sample data
my_portfolio_return_df.head()
```

## Join your portfolio returns to the DataFrame that contains all of the portfolio returns

```python
# Join your returns DataFrame to the original returns DataFrame
final_return_df = pd.concat([combined_return_df,my_portfolio_return_df],axis="columns",join="inner")
# Display sample data
final_return_df.head()
```

```python
# Only compare dates where return data exists for all the stocks (drop NaNs)
final_return_df.isnull().sum() 
#there is no NaNs in the dataset as only common dates have been selected due to inner join
```

```python
# Plot daily returns of all portfolios
final_return_df.plot(xlabel="date",ylabel= "daily return",title="Daily Return Comparison", figsize=(14,7));
```

```python
# Calculate cumulative returns of all portfolios
cum_final_return_df = (1 + final_return_df).cumprod()
# display few records
cum_final_return_df.head()
```

```python
# Plot cumulative returns
cum_final_return_df.plot(xlabel="date",ylabel= "cumulative return",title="Cumulative Return Comparison", figsize=(14,7));
```

## Re-run the risk analysis with your portfolio to see how it compares to the others

### Calculate the Annualized Standard Deviation

```python
# Calculate the annualized `std`
final_return_df.std() * np.sqrt(252)
```

### Calculate and plot rolling `std` with 21-day window

```python
# Calculate rolling standard deviation
rolling_21day_stddev = final_return_df.rolling(window=21).std()
# Plot the rolling standard deviation
rolling_21day_stddev.plot(xlabel="date",ylabel= "standard deviation",title="21-day Rolling Std Dev", figsize=(14,7));
```

### Calculate and plot the correlation

```python
corr_final_return_df = final_return_df.corr()
# Display de correlation matrix
corr_final_return_df.style.background_gradient()
```

### Calculate and Plot Rolling 60-day Beta for Your Portfolio compared to the S&P 500

```python
# Calculate covariance of a single portfolio
my_port_covariance = final_return_df['MY_PORTFOLIO_RETURN'].rolling(window=60).cov(final_return_df['sp500_return'])
# Calculate variance of S&P 500
variance = final_return_df['sp500_return'].rolling(window=21).var()
# Computing beta
my_port_beta = my_port_covariance/variance
my_port_beta
# Plot beta trend
my_port_beta.plot(xlabel="date",ylabel= "Beta",title="My Portfolio 60-day Beta",figsize=(14,7));
```

### Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot

```python
# Calculate Annualzied Sharpe Ratios
final_sharpeRatio = (final_return_df.mean() * 252) / (final_return_df.std() * np.sqrt(252))
final_sharpeRatio
```

```python
# Visualize the sharpe ratios as a bar plot
final_sharpeRatio.plot(kind="bar",xlabel="portfolio",ylabel= "sharpe ratio",title="Sharpe Ratio Comparison", figsize=(14,7));
```
